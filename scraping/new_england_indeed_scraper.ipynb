```python
# This is a separate block because sometimes the jobspy installation is malformed.
from jobspy import scrape_jobs
import pandas as pd
```

```python
def get_jobs(site:str, states:list[str], amount:int) -> pd.DataFrame:

    country_indeed='USA'
    results_wanted=1000 # 1000 is the max for indeed https://github.com/speedyapply/JobSpy?tab=readme-ov-file (see Notes section in readme before FAQ)

    frames = []
    for state in states:

        state_jobs = scrape_jobs(
            site_name=site,
            location=state,
            google_search_term=state,
            country_indeed=country_indeed,
            results_wanted=results_wanted,
            linkedin_fetch_description=True
        )
        frames.append(state_jobs)

    concat = pd.concat(frames)
    return concat
```

```python
new_england_states=['Maine', 'Massachusetts', 'New Hampshire', 'Rhode Island', 'Vermont', 'Connecticut']
ne_indeed_jobs = get_jobs('indeed', new_england_states, 10000)
```
