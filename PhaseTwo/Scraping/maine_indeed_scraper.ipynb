```python
from jobspy import scrape_jobs
import pandas as pd
```

```python
site_name=['indeed']
new_england_states=['Maine', 'Massachusetts', 'New Hampshire', 'Rhode Island', 'Vermont', 'Connecticut']
country_indeed='USA'
results_wanted=1000 # This seems to be the max https://github.com/speedyapply/JobSpy?tab=readme-ov-file (see Notes section in readme before FAQ)

frames = []
for new_england_state in new_england_states:

    state_jobs = scrape_jobs(
        site_name=site_name,
        location=new_england_state,
        country_indeed=country_indeed,
        results_wanted=results_wanted
    )
    frames.append(state_jobs)

concat = pd.concat(frames)
concat
```
