```python
import pandas as pd
from tqdm import tqdm
tqdm.pandas()
```

```python
df = pd.read_csv('new_england_indeed_jobs.csv')
```

```python
def remove_extra(row):
    skills = row['skills']
    skills = skills.lower()

    end_dot = len(skills)
    try:
        end_dot = skills.index('.')
    except ValueError:
        pass # . not found

    end_newline = len(skills)
    try:
        end_newline = skills.index('\n')
    except ValueError:
        pass # \n not found

    end = min(end_dot, end_newline)

    # Only truncate if a delimiter was found (i.e., end is not len(skills))
    if end < len(skills):
        skills = skills[0:end].strip() # Use strip() to remove trailing whitespace
        print(skills)



    # The print statement below was causing confusion by showing untruncated string before return. Removed it.
    # print(skills)
    return skills
```

```python
df['skills'] = df.progress_apply(remove_extra, axis=1)
```

```python
df.to_csv('new_england_indeed_jobs.csv', index=False)
```

```python
skills_series = (
    df['skills']
    .dropna()
    .str.split(',')
    .explode()
    .str.strip()
)

skills_df = skills_series.value_counts().reset_index()
skills_df.columns = ['Name', 'Frequency']
```

```python
skills_df = skills_df[skills_df['Frequency'] > 1].reset_index(drop=True)
```
